```python
import pandas as pd

#AGREGAMOS EL DATA SET Y VISUALIZAMOS LA PRIMERAS 5 COLUMNAS PARA VALIDAR QUE TODO ESTA BIEN#

ruta_pib = r"C:\Users\rosal\Documents\GitHub\practica-1-EDA-pj2r98\DATA\nama_10_gdp$defaultview_linear_2_0.csv"

df_pib = pd.read_csv(ruta_pib)
df_pib.head()
```

```python
# VALIDAMOS LA INFO QUE HAY DEL DATA SET (u estructira al igual que el tipo de dato como el valor de nulos que posee)#

df_pib.info()
print("\nDescripción general (primeras filas):")
```

```python
# VALIDAMOS TODAS LA COLUMNAS#

df_pib.columns
```

```python
#REVISAMOS VALORES UNICOS PARA SABER QUE FILTRAR#
df_pib['na_item'].unique()
df_pib['unit'].unique()
df_pib['freq'].unique()
df_pib['geo'].unique()

print(df_pib['na_item'].unique())
print(df_pib['unit'].unique())
print(df_pib['freq'].unique())
print(df_pib['geo'].unique())
```

```python
#Empezamos con el filtrado de la data que nos interesa#

df_pib = df_pib[df_pib['na_item'] == "B1GQ"]   #FILTRAMOS PARA QUE NOS QUEDE SOLO EL PIB (GDP)
df_pib = df_pib[df_pib['unit'] == "CLV10_MEUR"] #FILTRAMOS PARA QUE NOS QUEDE MILLONES DE EUROS A PRECIOS CONSTANTES
df_pib = df_pib[df_pib['freq'] == "A"]         #FILTRAMOS PARA QUE NOS QUEDE SOLO FRECUENCIA ANUAL

df_pib.head()
```

```python
#SELECCIONAMOS LAS COLUMNAS CON LAS QUE NOS QUEDAREMOS#

df_pib = df_pib[['geo',
                 'Geopolitical entity (reporting)',
                 'TIME_PERIOD',
                 'OBS_VALUE',
                 'na_item',
                 'unit']]

df_pib.head()
```

```python
#PROCEDEMOS A TRANSCRIBIR EL NOMBRE DE LAS COLUMNAS SEGUN LA INFORMACION DEL SATASET DE LA PAGINA OFICIAL
#  https://ec.europa.eu/eurostat/cache/metadata/en/gov_10q_ggfa_esms.htm , A ESPA;OL#

df_pib = df_pib.rename(columns={
    'geo': 'Codigo de Pais',
    'Geopolitical entity (reporting)': 'Pais',
    'TIME_PERIOD': 'Periodo',
    'OBS_VALUE': 'PIB (Millones de Euros)',
    'na_item': 'Indicador',
    'unit': 'Unidad'
})

df_pib.head()
```

```python
#AHORA SI PROCEDEMOS A VER NULOS Y EMPEZAMOS CON LA LIMPIEZA ANTES DE EMPEZAR CON TRANSFORMACION#

df_pib.isnull().sum()
```

```python
# REVISAMOS A;OS UNICOS DEL DATASET ORIGINAL#

df_pib['Anio'] = df_pib['Periodo'].astype(int)

print("Años disponibles en el dataset:")
print(df_pib['Anio'].unique())

print("\nCantidad de años:", len(df_pib['Anio'].unique()))
```

```python
# CREAMOS TABLA PIVOT FINAL (PIB por País y Año)

tabla_pib = df_pib.pivot_table(
    index='Pais',               # Filas → Países
    columns='Anio',             # Columnas → Años
    values='PIB (Millones de Euros)', # Valores numéricos
    aggfunc='sum'               # Como ya es anual, NO se promedia
)

# Mostrar tabla completa
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 2000)

print("\nTabla País vs Año (PIB Anual):")
print(tabla_pib)
```

```python
# EXPORTAMOS TABLA PIB A CSV
df_pib.to_csv(r"C:\Users\rosal\Documents\GitHub\practica-1-EDA-pj2r98\DATA\PIB_tabla_anual.csv",
              index=False,
              encoding="utf-8-sig")

print("Archivo CSV exportado correctamente.")
```
