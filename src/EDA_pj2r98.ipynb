```python
#AGREGAMOS EL DATA SET Y VISUALIZAMOS LA PRIMERAS 5 COLUMNAS PARA VALIDAR QUE TODO ESTA BIEN#

df = pd.read_csv("DATA/gov_10q_ggfa_page_linear_2_0.csv")
df.head()
```

```python
#Empezamos con el filtrado de la data que nos intersa#

df = df[df['na_item'] == "BF90"] #FILTRAMOS PARA QUE NOS QUEDE SOLO FQUE INDICAN LAS VARIABLES FINANCIERAS DEL GOBIERNO9#
df = df[df['sector'] == "S13"] #FILTRAMOS PARA QUE NOS QUEDE SOLO S13 QUE INDICA EL GOBIERNO EN GENERAL#
df = df[df['unit'] == "MIO_EUR"] #FILTRAMOS PARA QUE NOS QUEDE MILLONES DE EUROS#
df = df[df['finpos'] == "LIAB"] #FILTRAMOS PARA QUE NOS QUEDEN LOS PASIVOS#

df.head()
```

```python
#AHORA VAMOS A REDUCIR LAS COLUMNAS QUE SI NOS INTERESAN,
#YA QUE MUNAS DE ELLAS NOS SON IRRELEVANTES DE LAS CUALES NOS QUEDAMOS CON 6 DE LAS FILTRADAS#
```

```python
#SELECCIONAMOS LAS COLUMNAS CON LAS QUE NOS QUEDAREMOS#

df = df[['geo',
         'Geopolitical entity (reporting)',
         'TIME_PERIOD',
         'OBS_VALUE',
         'na_item',
         'finpos']]

df.head()
```

```python
#PROCEDEMOS A TRANSCRIBIR EL NOMBRE DE LAS COLUMNAS SEGUN LA INFORMACION DEL SATASET DE LA PAGINA OFICIAL
#  https://ec.europa.eu/eurostat/cache/metadata/en/gov_10q_ggfa_esms.htm , A ESPA;OL#

df = df.rename(columns={
    'geo': 'Codigo de Pais',
    'Geopolitical entity (reporting)': 'Pais',
    'TIME_PERIOD': 'Periodo',
    'OBS_VALUE': 'Millones de Euros',
    'na_item': 'indicador',
    'finpos': 'Posicion Financiera'
})

df.head()
```

```python
# CREAMOS COLUMNAS Y TRIMESTRES#
df['Anio'] = df['Periodo'].str.slice(0, 4).astype(int)
df['Trimestre'] = df['Periodo'].str.slice(5)

# MOSTRAMOS EL RESULTADO POR PAIS#
print("Primeros registros con País, Año y Trimestre:")
print(df[['Pais', 'Periodo', 'Anio', 'Trimestre']].head(20))
```

```python
# Crear columnas Año y Trimestre
df['Anio'] = df['Periodo'].str.slice(0, 4).astype(int)
df['Trimestre'] = df['Periodo'].str.slice(5)

# TABLA PIVOT
tabla_trimestres = df.pivot_table(
    index='Pais',
    columns='Trimestre',
    values='Millones de Euros',
    aggfunc='mean'
)

# MOSTRAMOS LA TABLA COMPLETA
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 2000)

print("Tabla País vs Trimestres COMPLETA:")
print(tabla_trimestres)
```

```python
#A PARTIR DE LA TABLA ANTERIOR CREAMOS UN NUEVO DATAFREME#
```

```python
df_trimestres = tabla_trimestres.copy()

print(df_trimestres)

```

```python
#CREAMOS LA NUEVA COLUMNA PROMEDIO#

df_trimestres['Promedio_Anual'] = df_trimestres[['Q1', 'Q2', 'Q3', 'Q4']].mean(axis=1)

#MOSTRAMOS TODA LA TABLA#
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 2000)

print("Nuevo DataFrame con Promedio Anual:")
print(df_trimestres)
```

```python
# REVISAMOS A;OS UNICOS DEL DATASET ORIGINAL#
df['Anio'] = df['Periodo'].str.slice(0, 4).astype(int)

print("Años disponibles en el dataset:")
print(df['Anio'].unique())

print("\nCantidad de años:", len(df['Anio'].unique()))
```

```python
# EXPRAEMOS EL A;O DESDE EL PERIDODO #
df['Anio'] = df['Periodo'].str.slice(0, 4).astype(int)

# CREAMOS NUEVA TABLA PIBOT País vs Años (media anual)
tabla_anual = df.pivot_table(
    index='Pais',              # filas = países
    columns='Anio',            # columnas = años
    values='Millones de Euros',# valores numéricos
    aggfunc='mean'             # promedio anual
)

# MOSTRAMOS LA TABLA COMPLETA #
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 2000)

print("\nTabla País vs Años (Promedio Anual):")
print(tabla_anual)
```

```python
#EXPORTAR TABLA ANUAL EN CSV
tabla_anual.to_csv("NETW_tabla_anual.csv", index=True, encoding="utf-8-sig")
```
