```python
import pandas as pd

# Traemos las rutas de tus archivos
path_netw = r"C:\Users\rosal\Documents\GitHub\practica-1-EDA-pj2r98\NETW_tabla_anual.csv"
path_pib = r"C:\Users\rosal\Documents\GitHub\practica-1-EDA-pj2r98\PIB_tabla_anual.csv"
path_pas = r"C:\Users\rosal\Documents\GitHub\practica-1-EDA-pj2r98\Pasivos_tabla_anual.csv"


```

```python
# AQUI CARGAMOS LOS DATAFRAMES PARA EMPEZAR A TRABAJAR CON ELLOS
df_netw = pd.read_csv(path_netw)
df_pib  = pd.read_csv(path_pib)
df_pas  = pd.read_csv(path_pas)

# AQUI REVISAMOS SUS COLUMNAS PARA ENTENDER COMO VIENEN ESTRUCTURADOS
print(df_netw.columns)
print(df_pib.columns)
print(df_pas.columns)

```

```python
#  TRANSFORMAMOS LA TABLA NETW DE FORMATO ANCHO A FORMATO LONG PARA PODER UNIRLA CON LAS DEMAS
df_netw_long = df_netw.melt(
    id_vars='Pais',
    var_name='Anio',
    value_name='NETW'
)

#  CONVERTIMOS LA COLUMNA ANIO A NUMERICA PARA PODER FILTRAR Y HACER COMPARACIONES
df_netw_long['Anio'] = df_netw_long['Anio'].astype(int)

#  FILTRAMOS SOLO DESDE 2018 PARA MANTENER COHERENCIA EN LOS 3 INDICADORES
df_netw_long = df_netw_long[df_netw_long['Anio'] >= 2018]

#  VEMOS UN ADELANTO PARA CONFIRMAR QUE TODO ESTA ORDENADO
print(df_netw_long.head())
```

```python
#  NOS QUEDAMOS SOLO CON LAS COLUMNAS UTILES DEL PIB PARA HACER LA UNION
df_pib_clean = df_pib[['Pais', 'Anio', 'PIB (Millones de Euros)']].copy()

#  RENOMBRAMOS LA COLUMNA PARA DEJARLA MAS LIMPIA Y CONSISTENTE
df_pib_clean.rename(columns={'PIB (Millones de Euros)': 'PIB'}, inplace=True)

#  NOS ASEGURAMOS DE QUE ANIO SEA NUMERICO PARA PODER FILTRARLO BIEN
df_pib_clean['Anio'] = df_pib_clean['Anio'].astype(int)

# FILTRAMOS SOLO LOS AÑOS A PARTIR DE 2018 PARA SINCRONIZARLO CON NETW Y PASIVOS
df_pib_clean = df_pib_clean[df_pib_clean['Anio'] >= 2018]

#  REVISAMOS COMO VA QUEDANDO LA TABLA LIMPIA
print(df_pib_clean.head())
```

```python
# AQUI TRANSFORMAMOS PASIVOS DE FORMATO ANCHO A FORMATO LONG PARA QUE PUEDA UNIRSE SIN PROBLEMAS
df_pas_long = df_pas.melt(
    id_vars='Codigo de Pais',
    var_name='Anio',
    value_name='Pasivos'
)

# AQUI RENOMBRAMOS LA COLUMNA DE PAIS PARA QUE COINCIDA CON LAS OTRAS TABLAS
df_pas_long.rename(columns={'Codigo de Pais': 'Pais'}, inplace=True)

# AQUI CONVERTIMOS EL ANIO A ENTERO PARA PODER FILTRAR Y HACER OPERACIONES
df_pas_long['Anio'] = df_pas_long['Anio'].astype(int)

# AQUI FILTRAMOS SOLO DESDE 2018 PARA QUE TODO ESTE SINCRONIZADO ENTRE LOS 3 INDICADORES
df_pas_long = df_pas_long[df_pas_long['Anio'] >= 2018]

# AQUI MOSTRAMOS UN ADELANTO PARA CONFIRMAR QUE TODO VA QUEDANDO CORRECTO
print(df_pas_long.head())
```

```python
# AQUI UNIMOS LAS 3 TABLAS (NETW, PIB Y PASIVOS) EN UNA SOLA TABLA FINAL
df_final = (
    df_netw_long
    .merge(df_pib_clean, on=['Pais', 'Anio'], how='outer')
    .merge(df_pas_long, on=['Pais', 'Anio'], how='outer')
)

# AQUI REVISAMOS LAS PRIMERAS FILAS PARA VER QUE LA UNION ESTE FUNCIONANDO CORRECTAMENTE
print(df_final.head())

# AQUI VEMOS LA FORMA FINAL PARA CONFIRMAR EL NUMERO TOTAL DE FILAS Y COLUMNAS
print(df_final.shape)

# AQUI EXPORTAMOS EL ARCHIVO FINAL LISTO PARA ENTREGAR O VISUALIZAR
df_final.to_csv("Uniondata_NETW_PASIVOS_PIB_2018_2025.csv", index=False, encoding="utf-8-sig")

print("ARCHIVO FINAL GENERADO: Uniondata_NETW_PASIVOS_PIB_2018_2025.csv")
```

```python
#  RENOMBRAMOS LA COLUMNA ANIO PARA DEJARLA MAS FORMAL COMO AÑO
df_final.rename(columns={'Anio': 'AÑO'}, inplace=True)

print(df_final.head())
```

```python
# CREAMOS UNA NUEVA TABLA PIVOT PERO SOLO COMO EJEMPLO DE ORGANIZACION
df_pivot = df_final.pivot_table(
    index='Pais',
    columns='AÑO',
    values=['NETW', 'PIB', 'Pasivos'],
    aggfunc='mean'
)
```

```python
#  DAMOS UNA DESCRIPCION PERSONALIZADA DEL NUEVO DATASET FINAL
print("\nDESCRIPCION DEL DATASET UNIFICADO (2018–2025):")
print("ESTE DATASET INTEGRA LOS INDICADORES NETW, PIB Y PASIVOS PARA CADA PAIS")
print("LAS CIFRAS DE PIB ESTAN EN MILLONES DE EUROS")
print("LOS DATOS DE NETW , PASIVOS Y PIB ESTAN EN LOS VALORES PUBLICADOS POR EL ECB")
print("PIB = https://ec.europa.eu/eurostat/databrowser/view/nama_10_gdp/default/table?lang=en")
print("NETW = https://ec.europa.eu/eurostat/databrowser/view/gov_10q_ggfa__custom_18997631/default/table")
print("PASIVOS = https://ec.europa.eu/eurostat/databrowser/view/gov_10q_ggfa__custom_18987587/default/table")
print("AÑOS DISPONIBLES: 2018 A 2025")
print("TOTAL DE REGISTROS FINALES:", df_final.shape[0])

# MOSTRAMOS COMO VA QUEDANDO ESTA NUEVA TABLA PIVOTADA
print("\nPREVIEW DE LA NUEVA TABLA PIVOTADA (PAIS VS AÑO):")
print(df_pivot.head())
```

```python
# EXPORTAR TABLA ANUAL EN CSV
df_pivot.to_csv("UNIOUNDATANETW_PASIVOS_PIB.csv", index=True, encoding="utf-8-sig")
```

```python
# AQUI APLANAMOS LAS COLUMNAS MULTINIVEL PARA QUE EXCEL LAS LEA DE FORMA LIMPIA
df_pivot.columns = [f"{col[0]}_{col[1]}" for col in df_pivot.columns]

# AQUI RESETEAMOS EL INDICE PARA QUE PAIS VUELVA A SER UNA COLUMNA NORMAL
df_pivot = df_pivot.reset_index()

# EXPORTAMOS LA TABLA ORGANIZADA, PLANA Y PERFECTA PARA EXCEL
df_pivot.to_csv("UNIONDATA_NETW_PASIVOS_PIB_ORGANIZADO.csv", index=False, encoding="utf-8-sig")

print("TABLA ORGANIZADA EXPORTADA: UNIONDATA_NETW_PASIVOS_PIB_ORGANIZADO.csv")
```
